```python
#importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Reading dataset
googleplaystore = pd.read_csv('googleplaystore.csv')
googleplaystore
```

```python
googleplaystore = googleplaystore.dropna()
```

```python
googleplaystore = googleplaystore.drop("Unnamed: 0", axis=1)
```

```python
# BAR Plot
def category_counts(data):
    counts = data["Category"].value_counts()
    plt.figure(figsize=(15, 8))
    ax = sns.barplot(x=counts.values, y=counts.index, hue=counts.index, dodge=False)
    plt.title("Category")
    for i, count in enumerate(counts):
        ax.text(count, i, str(count), ha='left', va='center', fontsize=10, color='black')
    if ax.legend_:
        ax.legend_.remove()
    plt.savefig('Bar01.png')
    plt.show()
 

category_counts(googleplaystore)
```

```python
# Histogram Plot
def rating_distribution(data):
    plt.figure(figsize=(10, 5))
    ax = sns.histplot(data[["Rating"]], kde=True)
    plt.title("Distribution of Ratings")
    plt.xlabel("Rating")
    plt.ylabel("Frequency")
    if ax.legend_:
        ax.legend_.remove()
    plt.savefig('Hist.png')
    plt.show()

rating_distribution(googleplaystore)
```

```python
# BAR Plot
def most_reviewed_apps(data):
    data["Reviews"] = data["Reviews"].astype(int)
    review = data[["Reviews", "App"]].groupby("App").sum().sort_values(by="Reviews", ascending=False).head(10)
    plt.figure(figsize=(10, 8))
    sns.barplot(y=review.index, x=review["Reviews"], palette="YlGnBu_r")
    plt.title("Most Reviewed Apps")
    plt.xlabel("Total Reviews")
    plt.ylabel("App")
    plt.savefig('bar02.png')
    plt.show()
    

most_reviewed_apps(googleplaystore)
```

```python
#PIE Plot
def app_types(data):
    plt.figure(figsize=(8, 8))
    fig = data.groupby("Type")["Type"].count().plot(kind="pie", autopct='%1.0f%%', shadow=False, explode=(0, 0.1))
    fig.set_title("App Types")
    fig.set_ylabel("")
    plt.savefig('pie.png')
    plt.show()

app_types(googleplaystore)
```

```python
# Scatter Graph
def installs_vs_rating(data):
    plt.figure(figsize=(30, 10))
    sns.scatterplot(x=data["Installs"], y=data["Rating"], hue=data["Type"])
    plt.title("Installs vs Rating")
    plt.xlabel("Installs")
    plt.ylabel("Rating")
    plt.savefig('scatter.png')
    plt.show()

installs_vs_rating(googleplaystore)
```

```python
# Line Graph
def ratings_of_paid_apps():

    plt.figure(figsize=(17, 8))
    sns.lineplot(data=googleplaystore, x="Price", y="Rating")
    sns.scatterplot(data=googleplaystore,x='Price',y='Rating')
    plt.xscale('log')

    plt.title("Distribution of Ratings of Paid Apps")
    plt.xlabel("Price ($)")
    plt.ylabel("Rating") 
    plt.savefig('line.png')
    plt.show()
 

ratings_of_paid_apps()
```

```python
# BOX Plot
def box():
    plt.figure(figsize=(10, 8))
    sns.boxplot(googleplaystore['Rating'],showfliers=False)
    plt.xlabel('Category(Rating)')
    plt.ylabel('Rating(Value)')
    plt.savefig('Box.png')

box()
```
